```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

tab_url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tab_name = "Коефіцієнт народжуваності в регіонах України"

tab_set = pd.read_html(tab_url, match=tab_name)
tab = tab_set[0]

# Вивести перші рядки таблиці за допомогою методу head
print(tab)
```

```python
# Вивести перші рядки таблиці за допомогою методу head
print(tab.head())
```

```python
# Визначте кількість рядків та стовпців у датафреймі
print(tab.shape)
```

```python
# Замініть у таблиці значення "—" на значення NaN
import numpy as np

tab1 = tab.replace("—", np.nan)#.convert_dtypes(convert_integer=False, convert_boolean=False)
print(tab1)
```

```python
from pandas.core.dtypes.missing import isnull
### Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_percentage = tab1.isnull().sum() / len(tab1) * 100
print(missing_percentage)
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
tab2 = tab1.drop(27)
print(tab2)
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
tab3 = tab2.fillna({"1950": tab2["1950"].mean()
                   ,"1960": tab2["1960"].mean()
                   ,"1970": tab2["1970"].mean()
                   ,"1990": tab2["1990"].mean()
                   ,"2000": tab2["2000"].mean()
                   ,"2012": tab2["2012"].mean()
                   ,"2014": tab2["2014"].astype('float64').mean()
                   ,"2019": tab2["2019"].astype('float64').mean()
                    }, axis=0)
tab3
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
snr = tab3["2019"].astype('float64').mean()
print(f"Середня народжуваність у 2014 році = {snr} на 1000 осіб")
print("\nРегіони з рівнем народжуваності більше середньої у 2014 році.\n")
print(tab3[tab3["2019"].astype('float64') > tab3["2019"].astype('float64').mean()])
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
nnr = tab3["2014"].astype('float64').max()
print(f"Найвища народжуваність у 2014 році = {nnr} на 1000 осіб")
print("\n")
print(tab3[tab3["2014"].astype('float64') >= nnr])
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

reg = tab3["Регіон"]
data = tab3["2019"].astype('float64')

plt.barh(
    reg,
    data,
    height=0.9,
    color=["b", "r", "y", "g"],
)

plt.ylabel("Регіон", fontsize="small", color="midnightblue")
plt.xlabel("Кількість народжень на 1000 осіб", fontsize="small", color="midnightblue")
plt.title("Народжуваність в Україні у 2019 році", fontsize=15)
plt.show()
```
