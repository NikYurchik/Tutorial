```python
# З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)?
# Відповідь: "Publication Manual of the American Psychologic..."
tab1 = tab.loc[tab['year']==2015]
max_price_2015 = tab1['price'].max()
print(f"max_price_2015: {max_price_2015}")
tab1.loc[tab1['price']==max_price_2015]
```

```python
# Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? Відповідь: 20
tab2 = tab.loc[(tab['year']==2010) & (tab['genre']=='Fiction')]
count_fiction_2010 = tab2['genre'].count()
print(f"count_fiction_2010: {count_fiction_2010}")
```

```python
# Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь: 1
tab3 = tab.loc[(tab['year'].isin([2010,2011])) & (tab['user_rating']==4.9)]
count_4_9_2010_2011 = tab3['user_rating'].count()
print(f"count_4_9_2010_2011: {count_4_9_2010_2011}")
tab3
```

```python
# І насамкінець, давайте відсортуємо за зростанням ціни всі книги, які потрапили до рейтингу в 2015 році і коштують дешевше за 8 доларів (використовуйте функцію sort_values).
# Відповідь: Яка книга остання у відсортованому списку? Відповідь: 253 "Old School (Diary of a Wimpy Kid #10)"
tab.loc[(tab['year']==2015) &(tab['price'] < 8)].sort_values(by='price')
```

```python
# Для початку давайте подивимося на максимальну та мінімальну ціни для кожного з жанрів
# (використовуйте функції groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min).
# Не беріть усі стовпці, виберіть тільки потрібні вам
tab.groupby(['genre']).agg(max_price=('price','max'), min_price=('price','min'))

# Максимальна ціна для жанру Fiction? Відповідь: 82

# Мінімальна ціна для жанру Fiction? Відповідь: 0

# Максимальна ціна для жанру Non Fiction? Відповідь: 105

# Мінімальна ціна для жанру Non Fiction? Відповідь: 0
```

```python
# Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count).
# Не беріть усі стовпці, виберете тільки потрібні
tab4 = tab.groupby(['author']).agg(count_books=('name','count'))
tab4.sort_values(by='count_books', ascending=False)

# Відповідь: Якої розмірності вийшла таблиця? Відповідь: 248 rows × 1 columns

# Відповідь: Який автор має найбільше книг? Відповідь: Jeff Kinney

# Відповідь: Скільки книг цього автора? Відповідь: 12
```

```python
# Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean).
# Не беріть усі стовпці, виберете тільки потрібні
tab5 = tab.groupby(['author']).agg(avg_rating=('user_rating','mean'))
print(tab5.sort_values(by='avg_rating'))

print(f"min_avg_rating: {tab5['avg_rating'].min()}")

# Відповідь: У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt

# Відповідь: Який у цього автора середній рейтинг? Відповідь: 3.9
```

```python
# З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну
tab6 = pd.concat([tab4, tab5], axis=1)

# Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)
tab6.sort_values(by=['count_books','avg_rating'])

# Який автор перший у списку? Відповідь: Muriel Barbery
```
