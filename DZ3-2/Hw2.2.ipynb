```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
missing_percentage = tab.isnull().sum() / len(tab) * 100
print(missing_percentage)
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
ind = tab.columns[tab.isnull().sum() > 0]
ind = ind[1:].to_list()
print(ind)
tab1 = tab.drop(ind, axis=1)
tab1
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
missing_percentage = tab1.isnull().sum() / len(tab1) * 100
print(missing_percentage)
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
tab2 = tab1.dropna()
tab2
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
tab2.shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = tab2.loc[tab2['Язык.программирования']=='Python']
python_data
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
grouped = python_data.groupby(['Должность'])
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg
# і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
# python_data2 = grouped['Зарплата.в.месяц'].agg(['min','max','mean'])
# print(python_data2)
python_data2 = grouped.agg(min_salary=('Зарплата.в.месяц','min'), max_salary=('Зарплата.в.месяц','max'), avg_salary=('Зарплата.в.месяц','mean'))
python_data2
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(s):
  res = (s['min_salary'] + s['max_salary']) / 2
  # res = (s.min_salary + s.max_salary) / 2
  return res

# python_data3.apply(fill_avg_salary, axis=1)
# print(python_data3)
python_data3 = python_data2.copy()
python_data3['avg'] = python_data3.apply(fill_avg_salary, axis=1).values
python_data3
```
