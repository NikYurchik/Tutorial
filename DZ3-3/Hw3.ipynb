```python
# напишіть функцію гіпотези лінійної регресії у векторному вигляді;
def linear_regression(w: np.array, X: np.array):
  return w @ X.T
```

```python
# # створіть функцію для обчислення функції втрат у векторному вигляді;
# y = w0 + w1*x
# y = w*X
def cost(y: np.array, h: np.array):
  return np.mean((h - y)**2)/2
```

```python
def step_descent(w: np.array, X: np.array, y: np.array, learning_rate: float=0.01):
  m = len(y)
  h = linear_regression(w, X)
  return w - learning_rate * 1/m * X.T @ (h-y)
```

```python
def grad_descent(w: np.array, X: np.array, y: np.array, epsilon: float=0.00001, num_iters: int=1000):
    last_cost_f = 10000000
    for i in range(num_iters):
      h = linear_regression(w, X)
      w = step_descent(w, X, y)
      new_cost_f = cost(y, h)
      if i % (num_iters / 10) == 0:
        print(f"i = {i}")
        print(f"last_cost_f = {last_cost_f}")
        print(f"new_cost_f = {new_cost_f}\n")

      if abs(last_cost_f - new_cost_f) < epsilon:
        print(f"abs(last_cost_f - new_cost_f) < {epsilon}")
        break
      last_cost_f = new_cost_f

    print(f"new_cost_f = {new_cost_f}\n")
    return w
```

```python
class LinearRegression:
  def __init__(self, lr: float=0.001, thr: float=0.00001, n_epochs: int=100):
    self.lr = lr
    self.thr = thr
    self.n_epochs = n_epochs
    self.w = np.array([random.uniform(-5, 5), random.uniform(-5, 5)])

  def predict(self, X: list):
    return self.w@X.T

  def update_weights(self, X: np.array, y: np.array):
    m = len(y)
    h = self.predict(X)
    self.w = self.w - self.lr * 1/m * X.T @ (h-y)

  def fit(self, X: np.array, y: np.array, w: np.array=None):
    if w is not None:
      self.w = w.copy()
    last_cost_f = 10000000
    for i in range(self.n_epochs):
      self.update_weights(X, y)
      new_cost_f = cost(y, self.predict(X))
      if i % (self.n_epochs / 10) == 0:
        print(f"i = {i}")
        print(f"last_cost_f = {last_cost_f}")
        print(f"new_cost_f = {new_cost_f}\n")

      if abs(last_cost_f - new_cost_f) < self.thr:
        break
      last_cost_f = new_cost_f

    print(f"new_cost_f = {new_cost_f}\n")
    return self.w
```
