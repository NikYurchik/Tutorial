```python
import random
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
tab_url = "/content/drive/My Drive/Colab Notebooks/Housing.csv"
tab = pd.read_csv(tab_url)

tab.head(10)
```

```python
# напишіть функцію гіпотези лінійної регресії у векторному вигляді;
def linear_regression(w: np.array, X: np.array):
  return w[0] + w[1] * X

w = np.array([3, 2])
x = np.array([5, 3])
result = linear_regression(w, x)
print(f"linear_regression: {result}")
```

```python
# # створіть функцію для обчислення функції втрат у векторному вигляді;
def cost(w: np.array, X: np.array, y: np.array):
    m = len(X) * 2
    predict = linear_regression(w, X)
    res = np.mean((predict - y)**2)/2
    return res

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([2, 3, 4, 4, 5, 6, 6])
w = np.array([0.0, 0.0])

loss = cost(w, x, y)
print(f"Функція втрат cost: {loss}")
```

```python
def step_descent(w: np.array, X: np.array, y: np.array, learning_rate: float=0.01):
    m = len(X)

    grad_w0 = (1/m) * np.sum(linear_regression(w, X) - y)
    grad_w1 = (1/m) * np.sum((linear_regression(w, X) - y) * X)

    w[0] -= learning_rate * grad_w0
    w[1] -= learning_rate * grad_w1

    return w


x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([2, 3, 4, 4, 5, 6, 6])
w = np.array([0.0, 0.0])

w = step_descent(w, x, y)
print(f"Оновлені ваги: {w}")
```

```python
# знайдіть найкращі параметри для датасету прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;
X = tab[['area', 'bedrooms', 'bathrooms']]
y = tab['price']

regres = LinearRegression()
regres.fit(X, y)

w = regres.coef_
w
```

```python
# знайдіть ці ж параметри за допомогою аналітичного рішення;
def analit(X: np.array, y: np.array):
  # w = (X.T @ X)^(-1) @ X.T @ y
  w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
  return w
```

```python
tab1 = tab[['area', 'bedrooms', 'bathrooms']]
y = tab['price']
a = np.ones((len(y), 1))
X2 = np.concatenate((a, tab1), axis=1).astype(int)
X2
```

```python
w2 = analit(X2, y)
w2 = w2[1:]
w2
```
